```python
df1=pd.read_csv("C:/Users/rajpu/OneDrive/Desktop/b/SleepQual classification.csv")
df1
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
from sklearn.metrics import classification_report
features = df1.drop("subject", axis=1)
target = df1["label"]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(scaled_features , df1.label )

predicted_labels = knn.predict(scaled_features)

cm = confusion_matrix(target, predicted_labels)
```

```python
plt.imshow(cm, cmap="Blues")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.title("Confusion Matrix")
plt.show()
```

```python
accuracy = accuracy_score(target, predicted_labels)
f1 = f1_score(target, predicted_labels, average="macro")

accuracy
```

```python
f1
```

```python
print(classification_report(target, predicted_labels))
```

```python


pca = PCA(n_components=5)
reduced_features = pca.fit_transform(scaled_features)
knn.fit(reduced_features, target)

predicted_labels_reduced = knn.predict(reduced_features)

accuracy_reduced = accuracy_score(target, predicted_labels_reduced)
f1_reduced = f1_score(target, predicted_labels_reduced, average="macro")

print("Accuracy after  reduction:", accuracy_reduced)
print("F1-score after  reduction:", f1_reduced)
print(classification_report(target, predicted_labels_reduced))
```
