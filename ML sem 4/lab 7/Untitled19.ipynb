```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

```

```python
data = pd.read_csv("Item_dataset.csv",header=None)
data = data.apply(lambda x: x.tolist())
products = data.apply(lambda x: [i for i in x if type(i) == str])
products
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
print("\nTransformed Dataframe:")
print(df)
frequent_items = apriori(df, min_support=0.20, use_colnames=True)

print("\nFrequent Items:")
print(frequent_items)

rules = association_rules(frequent_items, metric="lift", min_threshold=0.60)

print("\nAssociation Rules:")
print(rules)
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python

association_rules_50 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
print("\nAssociation Rules (min_support=50%, min_confidence=50%):")
print(association_rules_50)

rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=0.65)
print("\nAntecedent and Consequent Supports (min_lift=0.65):")
print(rules_lift[['antecedents', 'consequents', 'support']])
```
