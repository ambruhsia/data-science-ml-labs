```python
#a
import pandas as pd
file_path = "C:/Users/rajpu/OneDrive/Desktop/b/Employee_list.csv"
data = pd.read_csv(file_path, index_col='Name')
print(data)
#b
print()
data = pd.read_csv(file_path, index_col='Name')
sorted_data = data.sort_values(by=['Age', 'Salary'], ascending=[True, False])
print(sorted_data)
#c
print()
data = pd.read_csv(file_path, index_col='Name')
sort_data = data.sort_values(by=['Age', 'Salary'], ascending=[True, False])
print(sort_data)
#d
print()
data = pd.read_csv(file_path, index_col='Name')
salary_column = data['Salary']
median_salary = salary_column.median()
mode_salary = salary_column.mode().iloc[0]  
skewness_salary = salary_column.skew()
kurtosis_salary = salary_column.kurtosis()

print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
print("Skewness of Salary:", skewness_salary)
print("Kurtosis of Salary:", kurtosis_salary)
#e
print()
data = pd.read_csv(file_path, index_col='Name')
salary_column = data['Salary']
cumulative_sum_salary = salary_column.cumsum()
cumulative_product_salary = salary_column.cumprod()
cumulative_data = pd.DataFrame({
    'Cumulative Sum Salary': cumulative_sum_salary,
    'Cumulative Product Salary': cumulative_product_salary
})
print(cumulative_data)
```

```python
profession_mapping = {
    'Engineer': 0.0,
    'Doctor': 1.0,
    'Teacher': 2.0
}
data['Profession'] = data['Profession'].replace(profession_mapping)
data['Name'] = data.apply(lambda row: f"{row.Name} - {row['Profession']}", axis=1)
print(data)
```

```python
salary_df = pd.read_excel("C:/Users/rajpu/OneDrive/Desktop/b/EmployeeQ4.xlsx", sheet_name='Sheet1')
median_salary = salary_df['Salary'].median()
salary_df['Salary'] = salary_df['Salary'].fillna(median_salary).replace(0, median_salary).replace('', median_salary)
bonus_df = pd.read_excel('C:/Users/rajpu/OneDrive/Desktop/b/EmployeeQ4.xlsx', sheet_name='Sheet2')
merged_df = pd.merge(salary_df, bonus_df, on='Name')
merged_df['Effective Salary'] = merged_df['Salary'] + merged_df['Bonus']
output_df = merged_df[['Name', 'Effective Salary']]
output_df.to_csv('EffectiveSalary.csv', index=False)
print(output_df)
output_df
```
