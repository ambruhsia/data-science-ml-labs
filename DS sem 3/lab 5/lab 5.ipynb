```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as seaborn
#question1:
company=pd.read_csv("C:/Users/rajpu/OneDrive/Desktop/b/company_sales_data.csv")
company
```

```python
commulative_profit= company["total_profit"].cumsum()
monthnumber= company["month_number"]
totalprofit=company["total_profit"]
```

```python
#q1
plt.xkcd()
seaborn.set_style("darkgrid")
seaborn.lineplot(data=company, x=monthnumber,y=commulative_profit,color="#AA336A",marker="*",markersize=10).set_title("Commulative Profit for all months")
```

```python
#q2
seaborn.set_style("whitegrid")
q2=seaborn.lineplot(data=company,x=monthnumber,y=totalprofit,color="red",linestyle="dotted",marker="o",lw=3,markerfacecolor="red")
q2.set(xlabel='Month Number', ylabel='Sold Units Number')
q2.legend(loc='lower right', shadow=True)
```

```python
#q3
seaborn.set_style("darkgrid")
for product in company.columns[1:-2]:  
    seaborn.lineplot(x='month_number', y=product, data=company, label=product, marker='o')

plt.title('Units Sold per Month for Each Product')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.show()
```

```python

plt.plot('month_number', 'facecream', data=company, marker='o', label='Face Cream')
plt.plot('month_number', 'facewash', data=company, marker='o', label='Face Wash')
plt.plot('month_number', 'toothpaste', data=company, marker='o', label='Toothpaste')
plt.plot('month_number', 'bathingsoap', data=company, marker='o', label='Bathing Soap')
plt.plot('month_number', 'shampoo', data=company, marker='o', label='Shampoo')
plt.plot('month_number', 'moisturizer', data=company, marker='o', label='Moisturizer')
plt.legend(loc='upper left')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Number of Units Sold per Month for Each Product')
plt.show()
```

```python
#q4
plt.figure(figsize=(10, 6))
seaborn.set(style="whitegrid")
seaborn.scatterplot(x='month_number', y='toothpaste', data=company, marker='o', color='red')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Monthly Toothpaste Sales')
plt.grid(True)
plt.show()
```

```python
#q5
total_sales = company.iloc[:, 1:-2].sum()
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Sales by Product (Previous Year)')

plt.axis('equal') 
plt.show()
```

```python
#Q2:
covid=pd.read_csv("C:/Users/rajpu/OneDrive/Desktop/b/case_time_series (2).csv")
covid
```

```python

district=pd.read_csv("C:/Users/rajpu/OneDrive/Desktop/b/district.csv")
district
```

```python

plt.figure(figsize=(12, 8))
seaborn.set_style("darkgrid")
seaborn.lineplot(x='Date', y='Daily Confirmed', data=covid, markersize=20,marker='*',lw=7, markerfacecolor="blue",label='Daily Confirmed')
plt.xlabel('Date', fontsize=26)
plt.ylabel('Daily Confirmed Cases', fontsize=26)
plt.title('Daily Confirmed Cases Over Time')
plt.xticks(range(len(covid['Date'])), covid['Date'], rotation=45)

plt.legend(loc='upper left', fontsize=18)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(6))
plt.tight_layout()
plt.show()
```

```python
seaborn.set(style="whitegrid")
seaborn.set_palette("pastel")
plt.figure(figsize=(7,7))
seaborn.lineplot(x='Date', y='Daily Confirmed', data=covid, marker='o', markersize=10, label='Daily Confirmed')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Daily Confirmed Cases', fontsize=14)
plt.title('Daily Confirmed COVID-19 Cases Over Time', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, color='black', alpha=0.6)
plt.xlim('11 February', '17 May')  
plt.ylim(0, 3000) 
plt.tight_layout()
plt.show()
```

```python
seaborn.set(style="darkgrid")
plt.figure(figsize=(7, 7))
seaborn.barplot(x='districtData/0/confirmed', y='districtData/0/district', data=district , palette='viridis')
plt.xlabel('Confirmed Cases',color="#AA336A", fontsize=20)
plt.ylabel('District',color="#AA336A", fontsize=20)
plt.title('District-wise Coronavirus Confirmed Cases', fontsize=20)
plt.xlim(0,1400)
plt.tight_layout()
plt.show()
```

```python
iris = seaborn.load_dataset("iris")

plt.figure(figsize=(8, 6))
seaborn.set_palette("pastel")
seaborn.barplot(x="species", y="sepal_length", data=iris)
plt.title("Mean Sepal Length for Each Species")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length")
plt.show()
```

```python
iris = seaborn.load_dataset("iris")
plt.figure(figsize=(10, 6))
seaborn.lineplot(data=iris['sepal_length'], color='blue', label='Sepal Length')
seaborn.lineplot(data=iris['sepal_width'], color='green', label='Sepal Width')
seaborn.lineplot(data=iris['petal_length'], color='red', label='Petal Length')
seaborn.lineplot(data=iris['petal_width'], color='purple', label='Petal Width')
plt.title("Change in Measurements Over Id")
plt.xlabel("id")
plt.ylabel("Measurements (cm)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
seaborn.scatterplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, hue="Species")
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
iris = seaborn.load_dataset("iris")
seaborn.set(style="ticks")
seaborn.pairplot(data=iris, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)
plt.show()
```

```python
iris = seaborn.load_dataset("iris")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
seaborn.boxplot(x="species", y="sepal_length", data=iris,palette="Set1")
plt.title("Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.subplot(1, 2, 2)
seaborn.boxplot(x="species", y="sepal_width", data=iris,palette="Set2")
plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.tight_layout()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
seaborn.boxplot(x="species", y="petal_length", data=iris,palette="Set3")
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.subplot(1, 2, 2)
seaborn.boxplot(x="species", y="petal_width", data=iris,palette="pastel")
plt.title("Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.tight_layout()
plt.show()
```
