```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import seaborn as sns
import scipy.stats as stats
from scipy.stats import binom
```

```python


p = 0.5

rv = stats.rv_discrete(name='success', values=([1, 0], [p, 1-p]))

mean = rv.mean()
variance = rv.var()
skewness = rv.stats(moments='s')
kurtosis = rv.stats(moments='k')

(mean, variance, skewness, kurtosis)
```

```python
p = 0.8

random_numbers = bernoulli.rvs(p, size=1000)

count_of_successes = np.sum(random_numbers)
count_of_failures = len(random_numbers) - count_of_successes

data = {
    'Outcome': ['Failure (0)', 'Success (1)'],
    'Count': [count_of_failures, count_of_successes]
}

sns.barplot(x='Outcome', y='Count', data=data,palette="pastel")
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Bernoulli Distribution (p=0.8)')
plt.show()
```

```python

n = 30
p = 0.3
successes = np.arange(13)
probability_distribution = binom.pmf(successes, n, p)

data = pd.DataFrame({'Number of Infected People': successes, 'Probability': probability_distribution})
sns.barplot(x='Number of Infected People', y='Probability', data=data)
plt.title('Probability Distribution of Finding Infected People (n=30, p=0.3)')
plt.show()

```
